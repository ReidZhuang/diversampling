```python
import diversampling as ds
```

```python
import pandas as pd
# read data
data = pd.read_csv('creditcard_csv.csv')
data = data.drop(columns=['Time'], axis=0)
data.Class = data.Class.str[1]
# data cleaning (数据dataframe转换成array并增加index，分离标签)
display(data.head(5))
arr = data.to_numpy()
arr[:, -1] = arr[:, -1].astype(int)
arr_x = arr[:, :-1]
arr_y = arr[:, -1]
```

# Brief Version

```python
x = ds.Sample_data(arr_x, decreasing_ratio_threshold=0.005)
clustering = ds.cluster_process(x, cl_upper_num=50, accum_wcss_drop_ratio_threshold=0.02)
y = ds.Sample_label(arr_y)
usampling = ds.under_sampling(x, y, clustering.cluster_label, under_sampling_multiplier=2, random_multiplier = 0.05, n_neighbors=None)
display(usampling.array_under_sampled_final)    #final sample set after under-sampling
usampling.final_stats()    #sample balance stats of final dataset
```

# Complete Version

```python
# Instantiation of sample's x
x = ds.Sample_data(arr_x, decreasing_ratio_threshold=0.005)
x.scaler    #data after scaling process 
```

```python
display(x.pca_table.head(5))    #pca process stats
print('pca_var_num is ', x.pca_var_num)    #final dimensionality number
```

```python
display(x.pca_array)    #data after pca process
x.pca_model    #pca model
```

```python
# Instantiation of sample's y(label)
y = ds.Sample_label(arr_y)
y.stats    #label stats summary
```

```python
# Instantiation of the unsupervised classification process
clustering = ds.cluster_process(x, cl_upper_num=50, accum_wcss_drop_ratio_threshold=0.02)
display(clustering.kms_table.head(5))    #kmeans process stats table
print('The proper cluster number is ', clustering.cluster_num)    #final unsupervised cluster number 
```

```python
display(clustering.cluster_label)    #cluster label for data
clustering.kms_model    #kmeans model
```

```python
clustering.cluster_stats    #cluster stats information
```

```python
# Instantiation of the under-sampling process
usampling = ds.under_sampling(x, y, clustering.cluster_label, under_sampling_multiplier=2, random_multiplier = 0.05, n_neighbors=None)
usampling.cat_stat_tab    #cluster across label groupby results
```

```python
usampling.array_merge    #sample data array with x, y, and cluster label
usampling.tar_cat_list    #a list of cluster no. with minority type of sample
usampling.free_cat_list    #a list of cluster no. without minority type of sample
usampling.array_usampled    #under-sampling outcome for data from clusters in tar_cat_list
usampling.array_random_usampled    #under-sampling outcome for data from clusters in free_cat_list
usampling.usampling_info_df    #CondensedNearestNeighbour under-sampling stats restults. This table will show while a under_sampling instance is initialized.
display(usampling.random_usampling_info_df)    #random under-sampling stats results
```

```python
display(usampling.array_under_sampled_final)    #final sample set after under-sampling
usampling.final_stats()    #sample balance stats of final dataset
```

### Neg-Pos sample rate raised from 0.173% to 18.974%
